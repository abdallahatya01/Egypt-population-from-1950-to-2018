
Data Visualization


```python
plt.figure(figsize=(12,6))
fig=px.bar(df,
       x='Year',
       y='Total',
       title='number of population between 1950-2018',
       color='Total')
fig.show('png')       
```

```python
fig=px.bar(df,
       x='Year',
       y=['Rural','Urban'],
       title='number of population in Rural & Urban between 1950-2018',
       color_discrete_sequence=['yellow', 'red'] )
fig.show('png')              
```

```python
fig=px.bar(df,
       x='Year',
       y=['Male','Femal'],
       title='population of male& female between 1950-2018',
       color_discrete_sequence=['red', 'blue'])
fig.show('png')       
```
